```python
import pandas as pd
from pandas_datareader.data import DataReader
import pandas_datareader as prd
from datetime import date
import matplotlib.pyplot as plt
```

```python
start = date(1962,1,1)
end = date(2016,12,31)
ticker = "DGS10"
data_source = "fred"
stock_data = DataReader(ticker, data_source, start)
stock_data.info()
```

```python
prd.get_data_fred("GS10")
prd.get_data_fred()
```

```python
series_name = "10-year Treasury"
stock_data = stock_data.rename(columns={ticker: series_name})
stock_data.plot(title=series_name)
plt.show()
```

```python
start = date(2000,1,1)
series = "DCOILWTICO" # West Texas Intermediate Oil Price
oil = DataReader(series, "fred", start)
ticker = "XOM" # Exxon Mobile Corporation
stock = DataReader(ticker,"fred", start)
data = pd.concat([stock[["Close"]], oil], axis=1)
data.info()
```

```python
data.columns = ["Exxon", "Oil Prices"]
data.plot(title="Combined Options")
plt.show()
```
